## PCA, LDA :차원 줄이기

```python
#pca를 통해 차원 줄이기
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
out_pca = pca.fit_transform(df[['sepal length (cm)',
                                     'sepal width (cm)',
                                     'petal length (cm)',
                                     'petal width (cm)']])
```

```python
df_pca = pd.DataFrame(data=out_pca, columns=['pca1','pca2'])
print(df_pca.head())
```

괜찮은 것 처럼 보이지만 타깃 혹은 레이블 열이 없음 -> 원래 데이터프레임에서 데이터를 절단해 열을 더해줌
열을 더해주는 것을 예측에 사용될 수 있는 PCA 데이터 프레임을 제공해 줌
그것을 플록하고 변환된 데이터가 2차원으로 플롯된 것처럼 보이는지 확인

```python
df_pca=pd.concat([df_pca, df[['target']]], axis=1)
print(df_pca.head())
sns.lmplot(x="pca1",y="pca2", hue="target", data=df_pca, fit_reg=False)
```

고차원 데이터를 쉽게 소화할 수 있는 두개의 차원으로 전환
pca의 목표는 데이터를 가장 큰 변동이 있는 방향으로 맞추는 것, PCA는 데이터의 가장 중요한 정보들을 무시
레이블은 사용되지 않음: 레이블을 포함했으면 더 나은 변환 벡터들을 뽑아낼 수 있음
레이블을 포함한 차원 감소 기법은 선형판별분석(LDA)
LDA는 클래스 레이블별로 그룹을 지정, 클래스별로 가장 많은 분리를 생성하는 방향으로 찾음
** 번환 단계에서 레이블을 생략하는 것 : 도움이 되지 않는 방향으로 요소 벡터를 줄이는 것을 방지하기 위해 일부 문제의 경우(신뢰도가 낮은 클래스 레이블의 경우) 바람직함 -> 추가작업을 할지 말지를 결정하기 전에 항상 PCA부터 시작할 것을 추천, 데이터의 양이 방대하지 않다면 PCA부터 해도 나쁘지 않음

```python
# LDA를 이용해 차원수 줄이기
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=2)
```

```python
#데이터프레임 포맷 설정하기
out_lda=lda.fit_transform(X=df.iloc[:,:4], y=df['target'])
df_lda = pd.DataFrame(data=out_lda, columns=['lda1','lda2'])
df_lda=pd.concat([df_lda,df[['target']]],axis=1)
```

```python
sns.violinplot(x="target", y='lda1',data=df_lda).set_title("violinplot : Feature = LDA_1")
```

## 분리를 계량화 : K-평균 클러스터링과 실루엣 점수
클러스터의 압축성을 점수화하기 위해 실루엣 계수라 불리는 계량적 지표를 사용 : 높은 점수는 더욱 압축된 클러스터를 의미

```python
# K-평균으로 클러스터링하고 실루엣 점수 점검하기
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score as Sscore

#k-means 객체 초기화
kmns =KMeans(n_clusters=3, random_state=42)

#알고리즘을 pca에 적합시키고 실루엣 점수 산정
out_kms_pca = kmns.fit_predict(out_pca)
silhouette = Sscore(out_pca, out_kms_pca)
print("PCA silhouette_score = " + str(silhouette))

#알고리즘을 lda에 적합시키고 실루엣 점수 산정
out_kms_lda = kmns.fit_predict(out_lda)
silhoutte = Sscore(out_lda, out_kms_lda)
print("LDA silhouette score = %2f" %silhouette)
```

## 의사 결정 혹은 예측

예측을 하기 전에 데이터를 훈련 데이터와 테스트 데이터로 나눠야 함

훈련 데이터에 기반해 예측 모델을 수집하고 F1점수를 사용해 테스트 데이터를 점수화해봄

** 가장 랜덤화된 데이터 선택을 위해 랜덤 시드를 사용하는 것을 추천 : 어디서 랜덤화 루틴을 사용할지 의사 랜덤 생성기에 지정 -> 항상 같은 랜덤 선택이 제공


```python
# 데이터를 훈련/검증/테스트 그룹으로 나눔
from sklearn.model_selection import train_test_split as tts
df_train, df_test = tts(df_lda, test_size = 0.3, random_state=12)
```

```python
#데이터 점검
print('train set shape = '+ str(df_train.shape))
print('test set shape = '+ str(df_test.shape))
print(df_train.head())
```

```python
#SVM을 사용한 분류 : C=0.8
from sklearn.svm import SVC
from sklearn.metrics import f1_score
clf = SVC(kernel="rbf", C=0.8, gamma = 10)
clf.fit(df_train[['lda1','lda2']], df_train['target'])

#테스트 데이터상에서의 예측
y_pred =clf.predict(df_test[['lda1','lda2']])
f1=f1_score(df_test['target'], y_pred, average="weighted")

#예측 점수 점검
print("f1 score for SVM classifier = %3f" % f1)
```

모델을 바꿔서 적합할 수 있다

C는 패널티 항목이고 초매개변수라고 불림. 초 매개변수는 분석가가 적합을 특정한 방향응로 유도할 수 있는 항목

```python
#SVM을 사용한 분류 : C =1
from sklearn.svm import SVC
from sklearn.metrics import f1_score
clf = SVC(kernel="rbf", C=1, gamma = 10)
clf.fit(df_train[['lda1','lda2']], df_train['target'])

#테스트 데이터상에서의 예측
y_pred =clf.predict(df_test[['lda1','lda2']])
f1=f1_score(df_test['target'], y_pred, average="weighted")

#예측 점수 점검
print("f1 score for SVM classifier = %3f" % f1)
```

이렇게 매개변수를 조정해서 F1 점수를 최대화 : 매개변수를 자동화하기 위해 "그리드검색"을 적용할 수 있음

->대안은 튜닝을 필요로 하지 않은 알고리즘을 적용하는 것 : 랜덤포레스트 : 복수의 의사결정트리를 묶어서 예측에 대한 투표로 만드는 것 

```python
# 랜덤포레스트를 사용한 분류
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators =2, random_state=42)
clf.fit(df_train[['lda1','lda2']], df_train['target'])
y_pred =clf.predict(df_test[['lda1','lda2']])
f1 =f1_score(df_test['target'], y_pred, average='weighted')

#예측점수 점검
print("f1 score for SVM classifier = %2f" %f1)
```

#### 특징 스케일링
- 상수를 곱해도 성질이 바뀌지 않는다면, 그 수학적 성질은 '스케일 불변'으로 간주(ex-곡선의 모양이 입력의 강도와 상관없이 동일한 경우)
- 스케일링은 스케일 불변이 아닌 변환과 학습 알고리즘에 대해 중요하다.
- 주요 요소 분석(PCA), 페널티 회귀
- 문제를 수정하는 전략 은 각 특징을 비슷한 스케일로 적용, 분산과 편향에 집중, 가장 흔한 방법은 정규화와 표준화

#### 정규화
- 0과 1사이에 위치하도록 각 특징 공간을 리스케일하는 것
- MinMaxScalar() 함수에 있는 루틴 이용
- $x_{i,scaled}=\frac{x_{i,original}  -  min_{y}}{max_{y} - min_{Y}}$
    - $x_{i}$ = 데이터 포인트, $Y$ = x가 있는 열


변환을 이상치로부터 보호할 수 있는 방법은 없고, 이상치는 정규화 하기 전에 제거해야 한다., 그렇지 않으면 합당한 데이터들이 0-1 전체의 범위에 적절하게 펼쳐지니 못하고 유용한 데이터는 더 적은 범위로 모이게 됨

```python
# 모듈을 로딩하고 객체 로기화
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#데이터를 정규화하고 out_scaled numpy 행렬에 저장
out_scaled=scaler.fit_transform(df[cols])

out_scaled
```

#### 표준화

#### 순서적 인코딩

```python
#데이터에서 운동복 사이즈와 신발 사이즈 카테고리 변수를 알아보기
#카테고리 열에 필터 적용
cats = ['Jersey Size', 'Shoe Size']
df[cats]
```

 - Scikit-learn의 OrdinalEncoder 모듈을 순서적으로 열을 인코딩하기 위해 사용
 1. 항상 적절한 모듈을 임포트하고 인코더 객체를 예시하는 것부터 시작
 1. 데이터에 객체를 적합시키고, 한번에 .fit_transform()기법으로 변환
 1. 새로운 카테고리와 행렬을 .categroies_out_enc로 출력

```python
# 모듈을 임포트하고 enc 객체 초기화
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()

#한번에 객체를 적합 및 변환하고 카테고리 출력
out_enc = enc.fit_transform(df[cats])
print('identified categories : ')
print(enc.categories_)
print('encoded data : ')
print(out_enc)
```

#### 원핫 인코딩
- 바이너리 기법을 이용해서 기계의 상태를 기록하는 전자공학의 한 방법에서 출발
- 기계의 가능한 상태를 하나의 플래그로 정의, 현재의 상태와 맞는 플래그를 1로 하고 나머지 플래그를 0으로 하는 것
- 플래그는 원래 특징 열의 각 가능한 값에 대해 정의되고, 레코드가 그 상태에 있으면 1, 나머지 플래그는 0이 됨
- 단점 : 소스에서 가용하지 않은 새로운 상태를 외삽(extrapolate : 추정) 하는 것이 불가능
- 하는 법
    1. 모듈을 임포트, encoder 객체를 생성
    1. cats 리스트 내의 변수들을 적합, 변환
    1. scikit-learn의 .get_feature_next() 기법을 사용해서 새로운 열 이름을 불러오고 점검을 위해 출력

```python
# 모듈을 임포트하고 enc 객체를 초기화
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)

#한번에 객체를 적합 및 변환하고 카테고리를 출력
out_enc = enc.fit_transform(df[cats])
new_cols = enc.get_feature_names(cats).tolist()
print(new_cols)
```

```python
#원래 데이터와 연결된 임시 데이터 프레임 df_enc 생성
df_enc=pd.DataFrame(data = out_enc, columns = new_cols)
df_enc.index = df.index
df_enc
```

```python
#원래 열을 삭제하고 새로운 열과 연결시킴
df.drop(cats, axis=1, inplace=True)
df=pd.concat([df,df_enc],axis = 1)
print(df.columns)
```

#### 특징 필터링
- 목표가 중요성을 점수화하는 방법을 찾는 것이라면 가장 중요한 것만 보유하기
- 중요성을 점수화하는 방법은 여러가지가 있는데 가장 흔한 방법은 분산 한계점을 정의하고 상관계수에 의해 정렬하는 것

1. 분산한계점
- 한계점을 설정하고 그 값 아래에 있는 특징들을 제거하는 방법
- scikit-learn의 VarianceTreshold 객체는 적합할 때 각 특징의 분산을 저장하므로, 이를 활용하기 위해 사전 적합을 해보는 것을 추천
- 사전적합을 할 때는 모든 특징이 사전적합 후에도 남아있게 하기 위해 한계점을 설정하지 않음
    1. 한계점 없이 사전 적합
    1. 분산의 분석
    1. 한계점 선택
    1. 선택된 한계점을 가지고 재적합


```python
# iris 데이터 불러오기
df = pd.read_csv("./data/iris.csv")
df.index.name = "record"

#스케일링할 열 정의
cols = ['sepal length in cm',
        'sepal width in cm',
        'petal length in cm',
        'petal width in cm']
```

VarianceTreshold 객체를 생성, 한계점 없이 사전 적합한 후 분석

```python
# 한계점 없이 scikit-learn 객체 초기화
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold()

#df[cols] 사전 적합
selector.fit(df[cols])

#선택 이전에 특징 분산 점검
print(selector.variances_)
```

편의상 0.6을 한계치로 선택하고 재적합함. 결과에서 0부터 2까지의 열이 선택되는 것을 기대함

```python
# 선택 객체의 한계치 설정
selector.set_params(threshold=0.6)

#재적합 및 변환 후 out_sel에 결과 저장
out_sel = selector.fit_transform(df[cols])

#선택된 특징 점검
print(selector.get_support())
```

2. 상관계수
- 상관성 : 두 변수 간 일치성의 강도
- 피어슨 r 상관계수는 가장 흔히 사용되는 상관성 지표 [-1,1]
- -1,1에 가까우면 강한 상관관계, 0에 가까울수록 상관성이 낮음
- .corr() 기법과 seaborn의 히트맵을 사용해 시각적으로 복수의 r 상관계수를 한번에 분석 가능, 결과에서 가장 높은 상관관계를 갖는 특징을 필터링 할 수 있음

```python
# 컬러맵을 사요하기 위해 matplotlib 임포트
import matplotlib.pyplot as plt

#boston 데이터 로딩하기
from sklearn.datasets import load_boston
dataset = load_boston()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['MEDV'] = dataset.target
df.index.name = 'record'
```

pandas의 .corr() 기법을 사용해서 모든 변수의 쌍의 r 상관계수를 구함. seaborn 히트맵 루틴을 사용해서 상관계수를 플록할 수 있고, 컬러맵을 matplotlib의 Blues로 설정

```python
# pandas의 corr()을 사용해 상관계수 구하기
cor = df.corr()

#seaborn의 히트맵을 사용한 시각화
sns.heatmap(cor, annot=False, cmap=plt.cm.Blues)
plt.show
```

색이 진할 수록 더 강한 상관관계, 대각선은 자신과의 쌍이므로 r=1, 히트맵은 변수를 한번에 비교하거나 상관관계가 높은 변수를 찾는 경우 유용

```python
# 타깃 변수에 대해 상관계수 값 구하기
cor_target = abs(cor['MEDV'])
print(cor_target)
```

편의상 0.6을 한계점으로 하고 필터링, 결과에서 열 5와 12(0.69, 0.74)가 선택될 것을 기대함

```python
# 한계점 0.6 이상의 특징 선택
selected_cols = cor_target[cor_target>0.6]
print("selected columns, correlation with target > 0.6")
print(selected_cols)

#선택된 특징 필터 적용
df_sel = df[selected_cols.index]
print(df_sel.head())
```

#### 래퍼기법 wrapper
- 머신러닝 예측 알고리즘을 사용해서 좋은 예측을 위한 각 특징의 공헌도를 점수화
- 순차적 버전은 알고리즘 적합의 각 반복에서 특징 리스트를 업데이트함
- 업데이트는 전방 혹은 후방으로 할 수 있음
- 0개의 특징에서 시작해 하나 그 이상씩 선택하거나, 모든 특징에서 시작해 하나 혹은 그이상씩 뺄 수 있음

1. 순차적 특징 선택
- 전방, 후방으로 할 수 있음
- 분석가들이 이 용어를 사용할 때, 점수화 함수를 결정하고 전략을 업데이트 하는 루틴을 가르킴(특징을 더하거나 빼는 것)
- 그 루틴을 위해 커스텀 코드를 작성함
- 예측 알고리즘의 자동 점수화를 사용할 의향이 있다면, scikit -learn은 내장된 반복적 특징 제거 기법을 제공(RFE)

* RFE 기법은 적합 객체에 .coef_ 나 .feature_importance_ 속성이 있는 예측 알고리즘에서도 작동, 연속 타깃 변수에 대해서는 LinearRegression(), 카테고리 타깃 변수에 대해서는 RandomForestClassifier()을 추천

```python
# iris 데이터 불러오기
df = pd.read_csv("./data/iris.csv")
df.index.name = "record"

#스케일링할 열 정의
cols = ['sepal length in cm',
        'sepal width in cm',
        'petal length in cm',
        'petal width in cm']
```

서포트 벡터 머신 분류기(SVC)를 예제 RFE에서 추정기로 사용, 모듈을 임포트하고 SVC객체에 대해 독립변수(X)와 종속변수(Y)를 정의

```python
# RFE와 분류기 SVC에 대한 모듈 불러오기
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

#독립변수를 X로, 종속변수를 y로 지정
X=df[cols]
y=df['species']
```

RFE와 SVC 객체를 생성한 뒤 SVC 객체를 RFE의 인수로 전달. 결과 특징의 개수를 선택하기 위해 n_features_to_select 옵션을 사용

다음으로 적합 한 뒤 RFE ranking_ 속성을 사용해 특징 랭킹을 점검

```python
# RFE 객체를 생성하고 각 픽셀의 랭킹 생성
svc = SVC(kernel = "linear", C=1)
rfe = RFE(estimator = svc, n_features_to_select=2, step =1)
rfe.fit(X,y)

#랭킹 출력
print(cols)
print(rfe.ranking_)
```

선택된 속성들은 1의 중요성을 가짐, n_features_to_select="2"를 rfe 객체의 인수로 전달했기 때문, 객체는 두개의 특징을 선택 : 이 경우 petal length, petal width가 선택됨

#### PCA
- 비감독 방법, 데이터 차원을 감소시키기 위해 사용
- 새로운 특징 벡터를 파악, 데이터의 분산을 최대화 하고, 원래의 데이터를 새로운 공간에 투사하는 것
- 분산을 최대화 하는 새로운 특징 벡터 : 고유벡터 : 데이터의 주성분이 됨
- 원래는 특징당 하나의 성분이 있게 됨, 덜 중요한 성분은 버리고 가장 중요한 성분만 갖고 있어서 차원을 감소시키는데 유용
- scikit-learn은 각 주요 성분의 중요성에 대해 순위를 매길 때 사용될 수 있는 explained_variable_ 속성을 제공
- 주로 새롭게 정의된 낮은 차원을 쵸시하고 scikit-learn으로 그 차원을 분산으로 정렬하고 특징들을 제거하기 위해 n_components 옵션을 사용하게 됨

```python
# 두개의 결과 차원을 갖는 pca 객체 초기화
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

#두개의 입력 차원을 사용해서 적합 및 변환
out_pca = pca.fit_transform(df[['petal length in cm', 'petal width in cm']])

#pca 결과 데이터 프레임을 생성하고 'species' 레이블 명명
df_pca = pd.DataFrame(data = out_pca, columns=['pca1', 'pca2'])
df_pca = pd.concat([df_pca, df[['species']]], axis=1)

#pca 데이터 스캐터 플롯 생성
sns.lmplot(x='pca1', y='pca2', hue='species', data=df_pca, fit_reg=False)
```

- 각 주요 성분에 대한 설명된 분산은 급격히 하락
- 보통 50개, 100개의 차원이 있는 데이터에서 보통 2~3개의 주요 성분만 남김 : 보통 95% 이상의 분산을 설명하기 때문에
- pca1, pca 의 분산을 보여주기 위해 explained_variance_ratio 사용

```python
# 각 원소에 의해 설명되는 분산 구하기
print(pca.explained_variance_ratio_)
```

- 분리를 최대화 하는 새로운 틱징 벡터 : 판별식: 각 벡터는 자신과 자신 외의 모든 다른 벡터를 비교하는 판별식이 있음, n개의 벡터가 있는 경우 n-1개의 판별식이 있게 됨

### 그룹의 위치
#### 유클리디안 공간(센트로이드)
: 유클리디안 공간에서 숫자 데이터에 대해 클러스터의 위치를 파악하는 가장 흔한 방법은 각 데이터 포인트의 평균을 찾는것, 그 평균은 전체의 중심과 대응함 - 그 중심 : 센트로이드 : 기하학적인 중심이고 단순한 평균 계산으로 구할 수 있음

#### 비유클리디안 공간(메디오이드)
- 비 유클리디안 공간 : 문자열의 비교와 카테고리/숫자 데이터가 섞여있는 혼합 데이터
- 이런 경우 기하학적인 중심이 존재하지 않아 다른 전략 사용
- 가장 흔한 방법은 클러스터마다 메디오이드를 사용하는것
- 메디오이드(클러스토이드) : 클러스터 내의 다른 데이터 포인트와 가장 가까운 데이터 포인트
- 메디오이드는 실제 데이터 포인트 중의 하나가 되어야 함, 클러스터의 대푯값으로 간주될 수 있음

-> 다른 데이터와 가장 가까운 데이터 포인트는 무엇인가?
: 데이터 포인트 사이의 유사성 지표를 정함 -> 어떤 것이 다른 것들과 가장 가까운지 정하기 위해 이 유사성 지표에 기반해 모든 데이터 포인트를 비교

-> 어떻게 가까운 것을 정의하는가?
: 보통 다음 지표 중 가장 낮은 점수를 갖는 것을 의미
- 클러스터에서 다른 포인트 사이의 최대 거리
- 클러스터에서 다른 모든 포인트와의 평균 거리
- 모든 다른 포인트들까지의 거리 제곱합

> 선택된 지표와 상관 없이 메디오이드를 찾는 것은 상당히 노력이 필요, 클러스터 내의 모든 데이터 포인트가 비교돼야 하므로, 이 기법들은 쉽게 스케일링 되지 않음. 메디오이드의 계산 스케일은 $ O(n^2) $ 임

### 유사성
*데이터 마이닝에서 유사성 지표는 거리 지표라고도 불림

#### 유클리디안 공간
1. 유클리디안 거리 : 두 포인트 사이 최단 거리, L2표준
$ \sqrt{\sum_{i=1}^{10} (x_i - y_i)^2}$
2. 맨해튼 거리
: 택시지표, 도시블록거리, 두 포인트 사이의 거리, 주어진 그리드 모양을 따라 이동하는 거리, L1 표준
$\sum_{i=1}^{n} \vert x_i - y_i \vert $
3. 최대 거리
포인트 $x_i$와 다른 포인튼 $y_i $의 최대 거리
$ max_i ( \vert x_i - y_i \vert ) $

#### 비유클리디안 공간
- 좌표의 의미가 없음, 시스템의 축은 휘어졌거나 분리되어 있고, 공간에서 일관적으로 매핑될 수 없다. 
1. 코사인 거리
$ 1- \cos(\theta) $
2. 재커드 거리
: 두 개체에서 전체 속성 중에 공동의 속성 비율이 얼마인지를 구함
$ 1- \frac{X \cap Y}{X \cup Y} $

### 종료조건
- 여러개의 불확실한 부분이 있는 프로젝트를 시작하기 전에 종료조건이 무엇인지 확인하는 것이 좋은 습관
- 프로젝트가 완전히 종료되려면 무엇이 필요한가, 여기서 종료는 무엇을 의미하는 가
#### 알려진 숫자의 그룹일 경우
- 완료해야할 그룹의 개수가 알려진 경우 -> k개의 그룹 : k-평균 알고리즘
- 정확한 k의 개수가 알려졌으면 종료조건은 적절한 품질 점수의 수렴에 집중하게 됨
#### 알려지지 않은 숫자 그룹인 경우
- 생성되는 클러스터가 좋지 않은 클러스터로 판명될 때까지만 클러스터를 생성하는 수학적 루틴을 만드는 것

#### 품질점수와 실루엣
- 분석가들은 지름, 반지름, 밀도 계산을 품질 지표의 입력으로 사용
- 클러스터 내에서 모든 포인트의 전체 거리는 응집이라 부르고, 클러스터 사이의 전체 거리는 분리라 부름
- 가장 인기 있는 품질 점수는 실루엣 계수라고 부르고 응집과 분리의 균형을 의미합니다
- 포인트 $i$에 대해 $a$가 클러스터 안에서 다른 모든 포인트에 대한 평균거리이고 $b$가 가장 가까운 이웃 클러스트에 대한 평균 거리라면, 실루엣 계수는
- $S_i=\frac{(b_i) - (a_i)}{max(a_i , b_i)}$ / $max(a,b)$는 a,b 중에 더 큰 값을 취하는 것
- n개의 포인트에 대해 실루엣 점수는 : $mean(S_n)$
- 실루엣 점수는 -1에서 1 사이, 더 큰 점수는 더 나은 클러스터 품질을 의미

### 평균 분리
- 센트로이드에 따라 그룹을 나누고 수렴할 때까지 그들을 이동시키는 것
- 기법을 사용하기 위해서는 클러스터의 숫자(k)를 미리 알아야 함 : 알고리즘에 클러스터를 몇개로 할 것인지 지정해 줘야 함

#### k-평균
- 평균 분리의 기본 기법
- 단계
    - 입력 데이터 포인트에서 랜덤으로 k개의 초기 클러스터 중심을 고름
    - 모든 데이터 포인트를 가장 가까운 클러스터에 할당한다. 
    - k개의 센트로이드를 새롭게 생성된 클러스터 안의 중심으로 이동
    - k개의 클러스터가 더 이상 향상되지 않을 때까지 반복
- 위치를 정의하기 위해 센트로이드를 사용하고 유클리디안 거리를 유사성 지표로 사용하고 응집을 품질 점수로 사용
- 종료는 품질 점수가 수렴할 때, 즉 변화가 허용 오차 이내로 측정될 때 일어남
- k-평균 클러스터 객체를 초기화 할 때는 허용 오차를 tol 옵션을 사용해 설정, 많은 횟수의 적합 수행에도 허용 오차에 도달하지 않으면 max_iter을 사용해 강제 종료

```python
# 방울 예제 데이터 생성
df= make_blobs()

#모듈 임포트와 k-평균 객체 생성
from sklearn.cluster import KMeans
clus = KMeans(n_clusters=5, tol=0.004, max_iter=300,random_state=51)

#입력 데이터에 적합
clus.fit(df)

#입력 데이터 클러스터 할당과 처음 다섯개 레이블 출력
df['K-means Cluster Labels']=clus.labels_
print(df['K-means Cluster Labels'][:5].tolist())
df.head()
```

```python
#레이블이 표시된 그룹을 seaborn으로 시각화
sns.lmplot(x='Feature_2',y='Feature_1',
          hue="K-means Cluster Labels", data=df, fit_reg=False)
```

1. K 값 찾기
- 많은 데이터 마이닝 문제에서 k 값은 알려져 있지 않음
- k 값을 찾는 가장 흔한 방법은 적절한 품질 점수를 선택하고 가장 좋은 점수를 찾기 위해 k의 다른 값들을 적합해 보는 것(실루엣 점수 많이 사용)

```python
#실루엣 점수를 사용해서 최상의 k 값 선정
#metrics 모듈 임포트
from sklearn import metrics

#테스트를 위해 k 값의 리스트를 생성하고 for loop 사용
n_clusters = [2,3,4,5,6,7,8]
for k in n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(df)
    cluster_labels=kmeans.predict(df)
    S=metrics.silhouette_score(df,cluster_labels)
    print("n_clusters={:d}, silhouette score:{:1f}".format(k,S))
```

k=5 가 가장 좋은 선택으로 보임
> 데이터가 방대하면 k 값을 찾기 위해 데이터 일부만 적합할 수 있음, 이렇게 하면 적합을 찾기 위한 시간을 절약해 주어 더 많은 k 값을 시도해 볼 수 있음

2. k-평균++
- k-평균의 첫단계에서 랜덤으로 k개의 초기 클러스터 중심을 고르는데, 데이터가 방대할 경우 긴 적합시간을 유발할 수 있다, k-평균 알고리즘은 완전히 확정적이지 않고 글로벌 최솟값 대신에 로컬 최솟값에 수렴할 가능성도 있다. -> 개선한 것이 k-평균++
- k-평균++ 전략은 클러스터들이 멀리 떨어져 있다는 가정 하에 초기 클러스터 중심을 멀리 떨어져 있게 만듦

```python
#k-평균 init 기법을 사용해서 k-평균 객체 초기화
clus=KMeans(n_clusters=5, init="k-means++", tol=0.004, max_iter=300)
```

- 단! 적합 시간은 단축시킬 수 있지만, 글로벌 최솟값이 아닌 로컬 최솟값에 수렴할수도 있음

### 계층적 클러스터링
- 계층적인 방법으로 비슷한 클러스터를 합치는것
- 클러스터의 숫자는 어느 정도 수준의 계층이 문제에 적당한지에 따라 달라짐
- 계층적인 방법으로 연결된 클러스터의 플롯은 덴드로그램이라고 불림
    - 덴드로그램은 거리지표(y축), 데이터 포인트 기록(x축)로 나뉨
    - 그룹은 계층적으로 연결된 그래프에서 이뤄지므로 클러스터의 개수는 y축을 따라 위아래로 움직이면서 선택됨
    - 덴드로그램의 가장 높은 거리는 단일 클러스터가 낮고, 가장 낮은 것은 모든 데이터 포인트가 클러스터가 됨, 이 둘을 좋은 클러스터가 아니며 정답은 둘 사이 있어야 함
    - 가장 첫 단계는 거리행렬이라 부르는 행렬 만드는 것 
        - 연결행렬이라고 하는게 더 적절함
        - 행렬의 각 셀을 채우는 값은 쌍의 거리(연결)로 부터 옴
        - 연결은 전통적인 유사성 지표에 기반, 연결은 합쳐질 후보 클러스터를 비교하는 쌍의 지표가 됨
        - 연결 지표는 덴드로그램에서 y축이고, 낮은 값은 좀 더 비슷한 클러스터에 대응함
    - 계층적 클러스터링의 목표는 계층적인 방법으로 비슷한 클러스터를 결합하는 것임을 명심 -> 가능한 모든 클러스터를 결합 후보로 고려한다면, 가장 낮은 쌍의 연결을 갖는 두 클러스터가 결합하게 될 것
    - 인기 있는 결합 전략
        - 단일 연결 : 클러스터 [A]와 [B]에서 어는 두 포인트 사이의 최소 거리
        - 완전 연결 : 클러스터 [A]와 [B]에서 어느 두 포인트 사이의 최대 거리
        - 평균 연결 : 클러스터 [A]와 [B] 내의 모든 포인트 사이의 쌍 거리의 평균
        - 워드(ward) 연결 : 결합된 클러스터 [AB] 내의 모든 원소의 결과적 이득 
        
    - 제곱 오류의 합(SSE): 각 포인트에서 새롭게 생성된 센트로이드까지의 유클리디안 거리의 제곱합으로 정의-> 워드연결은 클러스터 내 포인트들에 얼마나 분산이 있는지 측정함으로써 각 클러스터를 점수화하게 됨
    - 계층적 클러스터링 분석 알고리즘(HCA)
        1. 응집적방법 : 모든 데이터 포인트가 각각의 클러스터이고 결합을 해나가면서 단일 클러스터 계층까지 올라가는 것 : 더 많이 쓰이는 방법
        2. 나눠지는 방법 : 단일 클러스터에 있는 모든 데이터 포인트로부터 시작해서 반대방향으로 내려감
    - 주요 옵션 : linkage 옵션을 사용해 연결 전략을 설정하고 affinity 옵션을 사용해서 연결에 쓰일 유사성 지표를 설정할 수 있음
    
        

```python
# 모듈 임포트와 HCA 객체 초기화
from sklearn.cluster import AgglomerativeClustering 
clus = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="ward")

#입력데이터에 대한 적합
clus.fit(df)

#클러스터 할당
df['HCA Cluster Labels']=clus.labels_

sns.lmplot(x="Feature_2", y="Feature_1",
          hue="HCA Cluster Labels", data=df, fit_reg=False)
```

#### 클러스터의 숫자를 찾기 위해 덴드로그램 재사용하기
- 분석하는 동안 덴드로그램은 한번만 생성하면 됨
- 알고리즘에 사용된 계층의 수를 거리 컷오프(덴드로그램에서 점선)를 위아래로 움직임으로써 변경할 수 있음
- 계층의 수는 클러스터의 수를 조정하므로, 클러스터의 수를 클러스터링의 품질을 조정하기 위해 사용할 수 있음,
- HCA 알고리즘은 $O(n^3)$으로 스케일링이 매우 나쁨 : 입력데이터 포인트 숫자의 세제곱으로 적합에 필요한 시간이 증가함을 의미
- 위 시간 복잡성은 덴드로그램의 구조와 관련되어 있으므로, 클러스터의 숫자 튜닝을 할 때는 다음 수행을 위해 덴드로그램을 저장하는 것이 좋음 ->재사용시 적합을 아주 빠르게 할 수 있음
- memory = "dir_to_store"과 compute_full_tree= True 옵션을 사용해서 적합 과정 동안에 지름길이 없이 전체 덴드로그램이 계싼되고 지정한 디렉터리에 저장되게 함

```python
#실루엣 점수를 사용해 클러스터의 최적 값 구하기
#metrics 모듈 임포트
from sklearn import metrics

#방울 예제 데이터 생성
df=make_blobs()

#모듈 임포트와 HCA 객체 초기화
from sklearn.cluster import AgglomerativeClustering 

#테스트할 k 값의 리스트 생성과 for loop의 사용
n_clusters=[2,3,4,5,6,7,8]
for num in n_clusters:
    HCA = AgglomerativeClustering(n_clusters=num, affinity = 'euclidean', linkage="ward", 
                                  memory='./model_storage/dendrogram', compute_full_tree=True )
    cluster_labels=HCA.fit_predict(df)
    S=metrics.silhouette_score(df, cluster_labels)
    print("n_clusters={:d}, silhouette score {:1f}".format(num,S))
```

5가 최적의 클러스터 숫자로 보여진다

### 밀도 클러스터링
- 유사성을 데이터 포인트 사이의 거리 지표를 유일한 지표로 삼는 대신, 밀도 클러스터링은 포이트에 의해 커버되는 공간을 추가 지표로 삼음
- 밀도 클러스터링에서는 주어진 공간에서 포인트 개수를 사용-> 잡음을 제거하는데 효과적 : 데이터의 밀집도가 높은 영역 밖의 이상치를 제거하기 쉬움
- 적합 루틴을 수행하기 전에 클러스터의 개수를 정할 필요가 없음
- 가장 인기 있는 밀도 클러스터링 알고리즘 : DBSCAN : 클러스터 안에서 데이터 포인트만 포함하도록 밀도의 정의를 제한하기 위해 응집 개념을 사용 = 포인트의 로컬밀도
- 유사성 지표 : 유클리디안 거리, 밀도 지표 : 클러스터 내 포인트의 수를 클러스터의 반지름($\epsilon$)으로 나눈 것, 더 나은 스케일링을 위해 $\epsilon$의 제곱이나 세제곱이 사용될 수도 있음
- DBSCAN 적합 루틴
    1. 분석가의 입력 : $\epsilon$ 값을 정의
    2. 이웃으로서 $\epsilon$에 의해 커벌 될 공간 고려
    3. 잡음으로 고려되지 않게 하기 위해 같은 이웃에 위치하게 될 최소의 다른 데이터 포인트 수(min_samples_를 정함
    : 이러한 입력을 이용해 다음의 레이블로 각 데이터 포인트를 할당
        - 코어 포인트 : 이웃 안에 min_samples 이상 있는 경우
        - 경계 포인트 : min_samples 요구 사항에는 미치지 못하지만, 코어 포인트의 이웃에 있는 경우
        - 잡음 포인트 : 코어 포인트 혹은 경계 포인트가 아닌 경우 
    4. 알고리즘은 서로의 이웃 안에 있는 코어 포인트를 결합함으로써 마무리 
    -> 결합된 코어 포인트들 : 최종 클러스터, 이웃 안에 있는 경계 포인트들은 그 클러스터에 할당, 잡음 포인트들은 클러스터에 할당 되지 않음(scikit-learn에서는 -1 값을 갖게 됨)

```python
# 방울 예제 데이터 생성
df=make_blobs()

#모듈 임포트와 DBSCAN 객체 초기화
from sklearn.cluster import DBSCAN
clus = DBSCAN(eps =0.9, min_samples=5, metric="euclidean")

#입력데이터에 적합
clus.fit(df)

#클러스터 할당
df['DBSCAN Cluster Labels'] = clus.labels_

sns.lmplot(x='Feature_2', y='Feature_1', hue='DBSCAN Cluster Labels', data=df, fit_reg=False)


```

이 경우 일반적인 K-평균 클러스터링이 더 나은 대안이다.  
DBSCAN이 모든 클러스터를 찾는데 실패했지만 잡음 포인트를 인지하는데 성공

### 스펙트럼 클러스터링
클러스터를 구성하는 노드의 연결성에 기반해서 연결 그래프를 생성하고 데이터 포인트를 그룹화함.  
밀도클러스터링과 달리 적합과정에서 클러스터의 수를 알아야 함  
- 유사성 행렬은 각 데이터 포인트와 나머지 포인트 간의 관련성을 비교
- PCA와 비슷하게 고유벡터를 차고 데이터는 새로운 관련성 공간으로 변환
- k-평균과 같은 전통적인 클러스터링 알고리즘이 관련성 공간에서 클러스터링을 위해 사용됨
- 유사성 행렬은 테이블의 각 원소가 쌍의 관련성 값($a_{ij}$)로 채워짐
최근접 이웃 알고리즘 : 이 알고리즘이 조정할 수 있는 자연스러운 초매개변수(이웃의 수)를 제공
> 계층적 클러스터링과 스펙트럼 클러스터링은 데이터의 상관관계 행렬을 정의하는 비슷한 접근방식처럼 보임. 하지만 매우 다른데, 계층적 클러스터링은 연결 거리에 의해 생성되는 연결행렬을 생성하고 그 거리에 기반해서 클러스터링을 함. 스펙트럼 클러스터링은 관련성 지표를 가지고 관련성 행렬을 생성. 그 다음으로 관련성 지표를 사용해서 클러스터링을 하는 대신, 클러스터링 이전에 저차원 고유벡터 공간으로 행렬을 변환시킴

<클러스터 객체에 대한 주요 옵션>  
affinity와 n_neighbor 옵션을 가지고 관련성과 이웃의 개수를 정한다.  
다음으로 assign_labels를 가지고 클러스터링 알고리즘을 선택하며, 선택적으로 k-평균 알고리즘을 한번 이상 수행하기 위해 n_init 옵션을 사용하고 최선의 적합을 택함

## 예측 개념 소개
미래 데이터에 대해 결과 값을 예측하는 것(회귀)이나 레이블을 예측하는 것(분류)는 데이터 마이닝 프로젝트에서 흔한 최종단계  
중요한 전략은 훈련 데이터를 수집하고 매핑함수(모델)를 입력 변수(x)와 결과 변수(y)에 대해 수립하는 것  
__<가정>__
- X와 y 사이에는 관계가 있음. x는 독립변수이고 y는 X에 의존함
- 미래 데이터는 훈련 데이터와 동일한 분포를 갖음
위의 두 가정이 성립되면, 훈련 데이터에 기반해서 모델을 수립하고 의미있는 예측을 생성하기 위해 새로운 데이터에 모델을 적용할 수 있음

매핑함수는 선형과 비선형 관계를 모두 모델링할 수 있고, 보통 최적의 적합을 위해 최적화될 수 있는 복수의 내부 변수를 가지고 있음(물론 매핑함수의 매개변수를 수동으로 바로 고르기를 원하지 않으므로, 최적의 매개변수를 찾는 함수를 만드는알고리즘을 설계해야 함)=> 컴퓨터가 매핑 함수를 학습할 수 있다. 

원하는 것을 계량적으로 설명할 수 있도록 수학을 사용해야 한다면, 수리적으로 정의해야 함
- 어떤 행동이 문제에 대해 중요한가
- 그 행동을 최적화할 수 있는 전략

가장 흔한 전략 : 예측 알고리즘을 최소화 문제로 만드는 것 
- 나쁜 행동을 정의하고 최소화함 
- 나쁜행동을 맞추지 못한 예측으로 정의되고 손실이라 불리는 잘못된 예픅과 정도를 특정하는 지표에 의해 계량화됨
- 손실을 계산하는 함수는 손실함수라 불리고 예측된 결과($y_{pred}$)를 실제 결과(y)와 비교
- 손실을 최소화 하는 방법 : 기울기 하강(가장 흔함) : 최소화의 방향 안에서 시스템을 이동시키기 위해 미분방정식을 사용하는 것


- 에측 알고리즘의 복수의 모델 배개변수들은 최소화 경로에 영향을 줄 수 있도록 지정될 수 있음 : 초매개변수- 최소화 문제와 별도로 지정됨
- 새로운 데이터에 대한 신뢰성 있는 일반화를 위한 예측모델의 초 매개변수를 생성하고 튜닝하는것 

### 예측모델 표기법
- X : 독립 입력변수의 행렬
- $x_i$ :X로부터 i번째 단일 레코드/행
- Y : 실제 종속 타깃 변수의 행렬
- $y_i$ : Y로부터의 i번째 단일 레코드/행
- $y_{pred}$ : 예측 타깃 변수
- m :X에서 레코드 수 
- n: X에서 특징의 수

### 수학적 도구

1. 손실함수 : 예측 오류를 계량화 하는 수식
- 모델 예측의 손실을 계산하는 수학적 표현 
- 손실 : 나쁜 행동 혹은 잘못도니 예측을 계량화하는 지표
- _예_ :선형회귀
    - 선형회귀는 X가 선형 함수를 사용해서 y에 매핑된다는 가설 사용
    - $y_{pred} = h_{\theta}(x) = \theta_{1}x + \theta_{0}$
    - 선형회귀에서 유용한 손실 함수는 훈련 데이터에서 예측 선분이 타깃 y와 얼마나 떨어져있는지를 계산
    - 예측선에서부터 실제 값까지 떨어져 있는 것 : 나쁜 행동($y_{i}-h_{\theta}$를 계량화-> 나쁜 행동의 합 = 시스템에서의 손실
    - $J(\theta)\propto \sum_{i=1}^{m}(y_{pred,i}-y_{i})^2 = C\sum_{i=1}^{m}(h_{\theta}(x_{i})-y_{i})^2$
        - 전체손실 J는 상수 C에 의해 곱해진 모든 파란 선(나쁜행동)의 제곱합 : 나쁜 행동의 모든 값이 양이 되게 하기 위해 제곱을 사용함 : 과도한 예측이나 부족한 예측이 같은 양의 나쁨으로 표현될 수 있게 해줌
        - 이제 나쁜 행동을 계량화할 유용할 손실 함수를 정의했고 나쁜 행동을 최소화 하며 됨
        - $J(\theta)$에 대한 공식에서 $\theta_{1}, \theta_{2}$ 를 편의상 단일 변수$\theta$로 합침
        
> 손실함수 == 비용함수
        

2. 기울기 하강 : 알고리즘 내에서 더 적은 손실의 방향으로 매개변수를 움직여서 예측 오류를 최소화하는 루틴
- 컴퓨터가 나쁜 행동(잘못된 예측)을 최소화하는 가설함수 $h_{\theta}$의 매개변수($\theta$)를 학습하게 하는 것
- 위 손실함수 그래프에서 극소점일 때가 가장 잘못된 예측을 최소화 하는 것 
- 미분 기울기 하강 업데이트 방정식 
    - $\theta_{updated} = \theta_{initial} - 스텝사이즈$
    - $\theta_{updated} = \theta_{initial} - \alpha(\frac{dJ(\theta)}{d\theta})$
- 학습비율 : 
    - 너무 큰 학습비율은 바닥을 지나치게 왔다갓다 하느라 시간 낭비 함
    - 너무 작은 학습비율은 업데이트 스텝이 너무 작아 손실함수 위를 왔닥 갔다 하는데 시간이 오래 걸리고 결국 시간을 낭비
- x를 데이터의 단일 행으로 표시하고 X를 복수의 행으로 이뤄진 행렬로 정의하는 예측표기와 비슷
- 미분은 단일 매개변수릐 변화 비율인 반면, 기울기는 향렬에 저장된 복수의 매개변수의 변화비율

### 품질 체계 적합하기
- 모델 튜닝 과정의 일부는 훈련 데이터의 예측 정확성 뿐만 아니라 헬드 아웃 데이터에 대해서도 같은 것을 측정하는 것
- 과적합 : 고분산, 과소적합 : 고편향 -> 적절한 적합 : 분산과 편향의 조화

## 회귀

### 회귀모델 예측 지표
- 평균 오류를 계량화 하는 지표 : 평균 제곱 오류(MSE) : $ MSE = \frac{1}{m}\sum_{i=0}^{m}(y_{pred,i} - y_{i})^2$
    - MSE는 스케일에 영향을 받을 수 있음(단점) : 큰 입력 값을 가지면 큰 MSE를 갖고 작으면 그 반대 : 단일한 문제를 다룰 때는 장애가 되지 않지만 복수의 케이스를 비교하기엔 무리
- 결정계수: $R^2 점수 = 1-(\frac{MSE}{\sigma^2})$

### 회귀 예제 데이터

```python
#모듈 임포트
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

#훈련/테스트 데이터로 나눠진 boston 데이터를 구하는 함수
def get_boston() :
    #boston 데이터 셋 불러오기
    dataset = load_boston()
    df= pd.DataFrame(dataset.data, columns=dataset.feature_names)
    df['MEDV']=dataset.target
    df.index.name='record'
    #훈련 데이터와 테스트 데이터로 나누기
    X_train, X_text, y_train, y_test = train_test_split(df.loc[:, df.columns != 'MEDV'], 
                                                        df['MEDV'], test_size =.33, random_state=42)
    return [X_train, X_text, y_train, y_test]
```

### 선형 회귀
- 선형 회귀 알고리즘은 데이터가 직선에 작 적합될 수 있다는 가정에서 출발한다. 
- $y_{pred} = h_{\theta}(x) = \theta_{1}x+\theta_{0}, J(\theta)\propto \sum_{i=1}^{m}(y_{pred,i}-y_{i})^2$
    - => $ J(\theta)=C \sum_{i=1}^{m}(h_{theta}(x_{i})-y_{i})^2=\frac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x_{theta})-y_{i})^2 $
    - 상수 C는 기울기 하강 과정에서 미분을 더 쉽게 계산할 수 있도록 삽입됐고, 그 값은 (1/2m)

```python
#모듈 임포트
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#moon 데이터 구하기
X_train, X_test, y_train, y_test = get_boston()

#회귀 객체 초기화 훈련 데이터에 대한 적합
clf=LinearRegression()
clf.fit(X_train, y_train)

#테스트 데이터에서의 예측과 y_test에 대한 예측의 점수화
y_pred = clf.predict(X_test)
r2 = r2_score(y_test, y_pred)
print('r2 score is = '+str(r2))
```

### 다변량 형태로 확장
- X도 벡터, 세타도 벡터
- $X=
\begin{pmatrix}
x_{0}^1 & x_{0}^2 & x_{0}^3 \\
x_{1}^1 & x_{1}^2 & x_{1}^3 \\
x_{2}^1 & x_{2}^2 & x_{2}^3 \\
\end{pmatrix}
$
-$\Theta =
\begin{pmatrix}
\theta_{[0,1]}^1\\
\theta_{[0,1]}^2\\
\theta_{[0,1]}^3\\
\end{pmatrix}
$
=>
- 선형회귀 가설 함수의 다변량 형태  : $h_{\theta}(X)=X\Theta $
- 선형 회귀 손실함수(제곱항은 전치행렬을 곱한 형태) : $J(\Theta)=\frac{1}{2m}(X\Theta -Y)^T (X\Theta -Y)$


### 패널티 회귀를 활용한 규칙화
수정항을 더해서 과적합을 수정하는 것 : 규칙화  
규칙화의 흔한 전략 : 패널티 항을 도입  
패널티 손실 = 손실함수 + 패널티 항
=> 다변량 선형 회귀 손실 함수 -> $J(\Theta)=\frac{1}{2m}(X\Theta -Y)^T (X\Theta -Y)$ + 패널티 항

1. 규칙화 패널티 
- 규칙화 패널티의 항은 모델의 매개변수 세타를 타깃으로
- 전략 : 어떤 단일 매개변수 세타도 적합을 독점하지 못함 
    - 매개변수가 너무 커지면 전테적으로 더 작은 매개변수를 학습하도록 모델에 패널티 부과
    - 적합으로부터 너무 작은 매개변수 세타를 완전히 제거해서 중요한 특징만 포함하도록 설계
    
    
- 다변량 공식에서 $\Theta$ 행렬은 각 j번재 특징열에 대한 $\theta$값으로 이뤄짐
- n개의 특징 열에 $\theta$의 절댓값의 합을 취해서 손실함수에 더해주면 L1 패널티라 불리는 다음의 값을 얻음
- L1 패널티 손실 = 손실함수 + $\lambda\sum_{j=1}^{n}\vert\theta_{j}\vert$
- L1 패널티를 갖는 선형 회귀는 라소lasso 회귀라고도 함
- 다변량 선형 회귀 손실함수는 L1 패널티를 갖는 라소 회귀 형태로 쓰여질 수 있음
    - $J(\Theta)=\frac{1}{2m}(X\Theta -Y)^T (X\Theta -Y)+ \lambda\sum_{j=1}^{n}\vert\theta_{j}\vert$ 
    
- 모든 세타의 제곱합을 손실함수에 더해주면 L2 패널티라불리는 다음의 값을 얻음
- L2 패널티 손실 = 손실함수 + $\lambda\sum_{j=1}^{n}\theta_{j}^2$
- L2 패널티를 갖는 선형 회귀는 리지ridge회귀 하고 함
- 다변량 선형 회귀 손실 함수는 L2 패널티를 갖는 리지회귀 형태로 다음과 같이 쓰여짐
    - $J(\Theta)=\frac{1}{2m}(X\Theta -Y)^T (X\Theta -Y)+\lambda\sum_{j=1}^{n}\theta_{j}^2$
    
- L1 패널티는 적합으로부터 작은 세타의 값을 갖는 특징을 완전히 제거
- L2 패널티는 어떤 항도 제거하지 않지만 세타의 제곱 단위로 스케일링 되어 세타 값이 커지면 더 강한 패널티를 부과

```python
###라소 회귀 ###
#모듈 임포트
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score

#moon 데이터 구하기
X_train, X_test, y_train, y_test = get_boston()

#분류기 객체 초기화와 훈련 데이터에 대한 적합
clf = Lasso(alpha=0.3)
clf.fit(X_train, y_train)

#테스트 데이터 상의 예측과  y_test에 대한 예측 점수화
y_pred =clf.predict(X_test)
r2=r2_score(y_test, y_pred)
print('r2 score is = %.3f' %r2)
```

```python
###리지 회귀 ###
#모듈 임포트
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

#moon 데이터 구하기
X_train, X_test, y_train, y_test = get_boston()

#분류기 객체 초기화와 훈련 데이터에 대한 적합
clf = Ridge(alpha=0.3)
clf.fit(X_train, y_train)

#테스트 데이터 상의 예측과  y_test에 대한 예측 점수화
y_pred =clf.predict(X_test)
r2=r2_score(y_test, y_pred)
print('r2 score is = %.3f' %r2)
```

## 분류
수리적 값을 예측하는 회귀 모델과 달리 입력 데이터를 카테고리 클래스 레이블로 예측
- 분류 모델은 참/거짓, 저위험/중위험/고위험, 동물의 종과 같은 레이블을 할당할 수 있음
- 회귀 모델은 주택가격, 멀리뛰기 거리, 홈런개수와 같은 결과를 예측

### 분류 예제 데이터

```python
#모듈 임포트
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

#모의 moon 데이터를 얻기 위한 함수
def get_moon_data():
    #데이터 모음을 생성하고 훈련 데이터와 테스트 데이터로 나누기
    X,y =make_moons(n_samples=150, noise=0.4, random_state=42)
    X_train, X_test, y_train, y_test = \
        train_test_split(X,y, test_size = .33, random_state=42)
    return [X_train, X_test, y_train, y_test]
```

### 분류 모델 예측의 지표
- 모델의 예측능력을 계량화하기 위해서는 점수화 시스템이 필요
- 정확한 예측의 비율을 측정하는 단순 정확도는 보통 충분하지 않고 중요한 모델 편향을 감출 수도 있음  
<다른 지표>
    - 정밀도 : 긍정적인 예에 집중하고 얼마나 모델이 긍정적인 점수를 잘 예측하는지
    - 재현율 : 부정적인 측면을 고려하고 얼마나 모델이 긍정적인 점수를 찾아내는지  
    업계에서는 무 지표가 모두 고려되는 하이브리드 점수 사용, $F_{1}$이라고 알려진 조화평균 공식 사용
<오차행렬>
- 예측 positive & 실제 positive : TP
- 예측 positive & 실제 negative : FP
- 예측 negative & 실제 negative : TN
- 예측 negative & 실제 positive : FN  
<지표점수>
- 정밀도 : $\frac{TP}{TP+TN}$
- 재현율 : $\frac{TP}{TP+FN}$
- F1 = 2 * $\frac{정밀도*재현율}{정밀도+재현율}$

### 복수 클래스 분류
두개를 초과하는 타깃 결과 레이블을 예측할 수 있는 능력을 갖듬. 이진 분류기는 다중 클래스 문제로 선천적으로 확장 가능하지 않음
1. 일 대 모두 전략
    1. 모두 다른 레이블에 대응하는 각 레이블에 대해 별도의 분류기를 생성함
    2. 다음으로 예측 시각에 모든 분류기에 새로운 데이터를 입력하고 가장 높은 점수를 갖는 레이블을 데이터 포인트에 할당  
    : 각 레이블에 대해 한 분류기를 생겅하고 각 분류기에 모든 데이터 포인트를 입력
2. 일 대 일 전략
    1. 모든 레이블의 쌍의 조합에 대해 별도의 분류기를 생성 : 각 분류기는 현재의 쌍의 조합의 데이터 포인트만 적합
    2. 예측 시각에 모든 분류기에 새로운 데이터를 입력하고 
    3. 각 결과 예측을 단일 투표로서 수집한 후 가장 많은 득표를 한 레이블을 데이터 포인트에 할당  
    : 각 레이블에 대해 하나 이상의 분류기를 생성  
        - 1이 레이블의 수이면, 일대일 전략 : 분류기의 수 = $\frac{l(l-1)}{2}$

### 로지스틱 회귀
- 로지스틱 회귀는 이진 분류 기법
- 가설 함수는 시그모이드 함수의 형태를 가짐
    - $y_{pred} = h_{\theta}(z) = sigmoid(z) = \frac{1}{1+e^{-z}}$
      - $z = \Theta^T X$
    - 시그모이드 함수는 S 형태를 띔-> 2진 분류에 적합한 형태
- 손실함수는 계단식 함수
    - 하나 이상의 체계를 가지고 있으며, 각각은 다른 형태를 가짐
    - 일반적인 형태 : $J(\theta)=loss(x_{i}, y_{i})$
    - 두개의 다른 체계는 $x_i$가 주어졌을 때 $y_i$의 실제 값에 의해 정의됨 : y=1이면 한 체계에 있고 y=0이면 다른 체계에 있는 것
    - 로지스틱 회귀 손실에 대한 계단식 함수 : $loss(x_i , y_i) = -log(h_{\theta}(y=1일때), log(1-h_{\theta})(y=0일때)$
    
  

```python
###로지스틱 회귀 분류###
# 모듈 임포트
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

#moon데이터 얻기
X_train, X_test,y_train, y_test = get_moon_data()

#분류 객체 초기화와 훈련 데이터에 대한 적합
clf=LogisticRegression(solver='lbfgs')
clf.fit(X_train, y_train)

#테스트 데이터상에서의 예측과 y_test에 대한 예측 점수화
y_pred = clf.predict(X_test)
f1=f1_score(y_test, y_pred)
print('f1 score is = '+str(f1))
    
```

### 서포트 벡터 머신
- 로지스틱 회귀 모델의 개념을 확장해서 클래스 사이의 최대 마진을 찾아내는 분류기를 생성
- 마진의 경계는 두 예측된 영역을 분리하는 결정 경계에 가장 가까운 데이터 포인트에 의해 정의됨 : 이 포인트들의 이름이 서포트 벡터
- 계단식 함수의 간단한 모형을 사용해서 SVM을 설명하기
    - $ loss(x_{i}, y_{i}) = loss_{y=1}(x_{i}, y_{i}), loss_{y=0}(x_{i}, y_{i}) $ 
    이 관계를 단일 표현으로 만들고 동시에 계단식 함수에 의해 정의된 체계에 기반한 행동을 보존해야 함
        1. 관계들을 더하고 최종표현을 만들때까지 비례적 부호를 사용함  
        $J(\theta)\propto loss_{y=1}(x_i , y_i) +  loss_{y=1}(x_i , y_i)$
        2. 목표는 계단식 함수를 보존하는 것 
            1. y=1일 때 y=1일 때의 손해부분만 활성화 되길 원함 = $loss_{y=1}$의 부분을 꺼놓을 방법을 찾아야함 : $loss_{y=0}$에 (1-y)를 곱해줌
            2. y=0 일때 : $loss_{y=1}$에 y를 곱해 줌
        3. 최종  
        $J(\theta)= -(y_i)loss_{y=1}(x_i , y_i) -(1-y_i)loss_{y=1}(x_i , y_i)$
    - $loss(x_i, y_i)$의 형태 : 힌지hinge 손실
        - 힌지 : on 영역에서 선형의 모양을 갖고 off 영역에서 0의 값을 갖는 선분
        - 모델 적합의 목표는 손실함수를 최소화하는 것 -> 0의 값을 갖는 부분은 최소 손실의 위치가 됨
        - 힌지 모양의 결과를 유리하게 사용하고 원하는 선분이 발생하도록 힌지 곡선을 위치시킬것  
        __SVM은 클래스 사이의 최대 마진을 찾기 위해 직접 타깃되는 분류기__
        - y=1과 y=0 예측 체계 사이에서 공유된 경계를 타깃팅하는 대신에 두 경계를 분리하고 각각을 0으로부터 멀리 떨어뜨림
        - y=1이면 J(z)가 1보다 클 때 Z 값이 0보다 크도록, y=0이면 J(z)가 -1보다 작을 때 사라지게함으로써 z가 0보다 작게하는 식으로  힌지 곡선을 위치시킴으로써 이 작업을 완성시킴  


#### C를 사용한 소프트 마진
모든 데이터가 커다란 마진으로 깨긋하게 분리되지는 않음 => 마진의 정의를 완화하는 것  
마진을 완화하면 좀 더 강력한 벡터머신을 생성하기 위해 몇몇 샘플을 재분류하거나 마진 안에 위치하도록 할 수 있음  
규칙화 기법은 scikit-learn에서 C 패널티 옵션을 사용해 적용  
낮은 C값은 완화된 마진 의미  

```python
### 서포트 벡터 머신 분류 ###
#모듈 임포트
from sklearn.svm import SVC
from sklearn.metrics import f1_score

#moon 데이터 얻기 
X_train, X_test, y_train, y_test = get_moon_data()

#분류 객체 초기화와 훈련 데이터에 대한 적합
clf = SVC(kernel = 'linear', C=0.5)
clf.fit(X_train, y_train)

#테스트 데이터 상에서의 예측과 y_text에 대한 예측 점수화
y_pred = clf.predict(X_test)
f1=f1_score(y_test, y_pred)
print('f1 score = '+str(f1))
```

#### 커널 트릭
- 전통적인 SVM은 이진, 선형 분류기 BUT 커널트릭을 활용해서 비선형 분류에도 확장될 수 있음
- 커널트릭은 가설함수의 정의를 새로운 비선형 매핑함수로 대체
- 가장 흔한 커널 선택 : 가우시안 커널, 반경 기반 함수를 사용
    - gamma 옵션에 의한 계수 : 가우시안 매칭 함수의 폭을 조정, 각 포인트가 마진 정의에 대해 갖는 영향력의 구sphere을 조정
    - 작은 gamma 값은 큰 반경과 낮은 비선형성을 의미 -> gamma를 작게 지정할 수록 SVM은 성형 커널 버전처럼 작동

```python
###가우시안 커널 분류를 사용한 SVM ###
# 분류 객체 초기화와 훈련 데이터에 대한 적합
clf=SVC(gamma=2, C=1)
clf.fit(X_train, y_train)

#테스트 데이터상에서의 예측과 y_text에 대한 예측 점수화
y_pred=clf.predict(X_test)
f1=f1_score(y_test, y_pred)
print('f1 score= '+str(f1))
```

### 트리기반 분류
- 예측을 하기 위해 단계적인 방법으로 의사결정 논리를 적용 
- 단순한 의사결정과정은 인간의 논리 흐름과 비슷
    - 다중클래스의 문제들을 특별한 전략 없이 처리할 수 있고
    - 수리적 입력이나 카테고리 입력을 모두 처리할 수 있음
- 카테고리 입력은 최상의 결과를 얻기 위해 인코딩 필요 : 원핫 인코딩 등
- 약점 : 너무 복잡하게 자랄 경우 과적합의 경향이 있음
    - 리오 브라이먼의 인사이트에 의해 해결: 각 트리와 특징의 깊이를 제한해서 각 트리가 약한(과소적합의) 모델을 학습하게 함 -> 각각의 약한 학습 트리가 앙상블을 생성하고 예측에 대해 투표하도록 함 => 스스로 튜닝하고 새로운 데이터에 대해 일반화가 잘되는 모델이 됨
    
#### 의사결정 트리
- 인간이 결정하는 것돠 비슷한 방법으로 일련의 논리트리를 생성 -> 이해할 수 있는 아주 투명한 의사결정 함수
- 소스 노드에서 모든 데이터를 가지고 시작하며, 데이터의 특정한 특징을 사용해 두개의 노드로 갈라짐  
__<Gini를 활용한 노드 분할>__  
- 의사 결정 트리 알고리즘은 어디서 어떻게 노드를 분할할지 학습해야함 : 노드의 분할은 불순도에 기반한 관계에 의해 이뤄짐
- 불순도 : 현재 노드의데이터에 대해 얼마나 예측이 신뢰할만한지 보여주는 지표 
    - (수학적) 현재 콘텐츠의 분포에 기반해서 어떻게 노드가 랜덤하게 선택된 데이터 포인트를 잘 분류할 수 있는지
    - 최소 불순도 : 0 : 노드 안의 모든 예들이 같은 클래스에 있을 때
- 어디에서 분할할지에 대한 결정은 입력 특징과 그 전체 범위를 돌아다니면서 어떤 분할 위치가 자식 노드에서 최소의 불순도를 갖는지에 따라 달림
- 불순도 Iml -> 노드의 분할 위치 = min$(Imp_{child1}+Imp_{child2})$
- 불순도의 계량화 : 지니지수
    - 현재 노드의 i개의 클래스의 리스트가 [1,2,3, ..., j]이고 그 노드에서 데이터 포인트를 랜덤으로 선택했을 때 클래스를 맞게 선택할 확률이 p라면
    - $Gini imp = 1-\sum_{i=1}^{j} p_{i}^2$

```python
###의사결정 트리 분류###
#모듈 임포트
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

#moon 데이터 얻기
X_train, X_test, y_train, y_test = get_moon_data()

#분류 객체 초기화와 훈련 데이터에 대한 적합
clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train, y_train)

#테스트 데이터 상에서의 예측과 y_test에 대한 예측을 점수화
y_pred = clf.predict(X_test)
f1 = f1_score(y_test, y_pred)
print('f1 score ='+str(f1))
```

__<배깅을 사용한 내장된 검증 기능>__  
- 약한 학습 트리 전략은 각 트리에 대해 선택된 데이터 포인트에 집중할 때 훌륭하다는 것이 명확해짐
- 일부는 배깅이라 불리는 전략에 의해 의도적으로 제외됨
- 배깅을 랜덤으로 샘플의 일부를 백bag에 담는 것이고, 나머지는 out-of-bag(OOB)가 됨. OOB 집합은 내부적인 검증데이터가 됨
- 각 트리는 예측을 비교할 OOB 데이터를 가짐 => 랜덤 포레스트 모델의 성능을 검증하기 위해 별도의 교차검증 전략을 사용하지 않아도 됨
> 배깅은 보통 대체(replacemenet)로 이뤄짐, 겹치는 데이터가 각 백에 허용되는 것을 의미하고 부트스트래핑이라고 불림. 분류에서 부트스트랩의 결과는 각 백에 인공적으로 더 많은 변종을 더하는 것이 됨
- scikit-learn에서 랜덤 포레스트 분류는 RandomForestClassifier 모듈을 사용해 이뤄짐
    - n_estimator : 원소의 수 정함
    - max_features : 각 노드 분할에서 가능한 특징들을 정함
    - max_depth : 트리의 깊이 제한
    

```python
### 랜덤 포레스트 분류###
#모듈 임포트
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

#moon 데이터 얻기
X_train, X_test, y_train, y_test = get_moon_data()

#분류 객체 초기화와 훈련 데이터에 대한 적합
clf = RandomForestClassifier(max_depth=4, n_estimators=4, max_features='sqrt', random_state=42)
clf.fit(X_train, y_train)

#테스트 데이터 상에서의 예측과 y_test에 대한 예측을 점수화
y_pred = clf.predict(X_test)
f1 = f1_score(y_test, y_pred)
print('f1 score ='+str(f1))
```

객체 초기화 과정에서 oob_scroe 옵션을 사용해서 검증을 위해 랜덤 포레스트 모델로 하여금 OOB 점수를 계산할 수 있음

```python
### 검증 데이터를 위한 OOB 사용 ###
# 분류 객체 초기화와 훈련 데이터에 대한 적합
clf = RandomForestClassifier(max_depth=4, n_estimators=10,
                             max_features='sqrt', random_state=42,
                            oob_score=True)
clf.fit(X_train, y_train)

#OOB를 사용한 예측 점수화
oob_score=clf.oob_score_
print('OOB score = '+str(oob_score))
```
